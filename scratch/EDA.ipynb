# EDA

## Imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Data

```python
data = pd.read_csv('./data/merged_data.csv')
data.drop(columns='Unnamed: 0', inplace=True)
data.head()
```

## Pairplot

```python
plt.figure(figsize=(8, 12))
sns.pairplot(
    data=data[
        [
            "wave_height",
            "dominant_period",
            "avg_period",
            "water_temp",
            "dominant_wave_direction_sin",
            "dominant_wave_direction_cos",
            "wind_speed",
            "gust_speed",
            "wind_direction_sin",
            "wind_direction_cos",
            "score",
        ]
    ],
    corner=True,
);
```

```python
data.corr()['score'].sort_values()
```
